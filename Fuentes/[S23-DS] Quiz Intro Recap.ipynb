**Rules to remember:**



*   As in Java, parameter **values** are copied to the local parameter 
variables  **in the order they appear**.


*   We can use the **local parameter variable names** in the calling function.

What is the output of the following cell?

```python
def f(a=3, b=6):
    return a+2*b

a=5
b=8
print(f(b))
```

What is the output of the following cell?

```python
def f(a=3, b=6):
    return a+2*b

a=5
b=8
print(f(b,a))
```

# **Binary Search Trees**

## **Fundamentals**

```python
class BST:

    def __init__(self, k=None):
        if (k is None): # if k is None, it will create an empty tree
            self.is_empty = True
        else:           # otherwise it will create a tree with a single node, containing k
            self.is_empty = False
            self.key=k
            self.left=BST()   # left  child will be an empty tree
            self.right=BST()  # right child will be an empty tree
```

The function **T.find(k)** returns the address of the node in the tree with root T where k is, or None if k is not in the tree.

If we set the parameter return_parent to True, the function also returns the address of the parent of the node where k is.


```python
T = bst.BST.from_list([5,3,1,8,4,2,9,7,6])

T.draw()

x = T.find(7)
print(x, x.key)

x = T.find(10)
print(x)

p, x = T.find(8,return_parent=True)
print(p, p.key)
print(x, x.key)

p, x = T.find(10,return_parent=True)
print(p)
print(x)
```

## **Problem Type 1**  - Function that traverses a single path that is independent of parameters and tree contents 

```python
def smallest(T):
  if T.left.is_empty:
    return T.key
  return smallest(T.left)

print(smallest(T))
```

## **Problem Type 2** - Function that traverses the whole tree, independently of parameters and tree contents 

```python
def bst_to_list(T):
  if T.is_empty:
    return []
  return bst_to_list(T.left) + [T.key] + bst_to_list(T.right)

print(bst_to_list(T))
```

## **Problem Type 3** - Function that traverses a single path that is dependent of parameters and tree contents 

```python
def depth_of_k(T,k):
  if T.is_empty:
    return -1
  if T.key ==k:
    return 0
  if T.key>k:
    d = depth_of_k(T.left,k)
  else:
    d = depth_of_k(T.right,k)
  if d<0:
    return -1
  return d+1

T.draw()
for k in range(12):
  print('Depth of',k,'is',depth_of_k(T,k))
```

## **Problem Type 1**  - Function that traverses a single path that is independent of parameters and tree contents 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
T = btree.BTree()
T.extend(np.random.permutation(30)*2)
T.draw()
```

```python
def min(T):
  if len(T.child)>0:
    return min(T.child[0])
  return T.data[0]

print('The minimum item in  the tree is',min(T))
```

## **Problem Type 2** - Function that traverses the whole tree, independently of parameters and tree contents 

```python
T.draw()
```

```python
def count_nodes(T):
  count = 1
  for c in T.child:
    count += count_nodes(c)
  return count

print('The number of nodes in the tree is',count_nodes(T))
```

```python
def items_in_order(T):
  if len(T.child)==0:
    return T.data.copy()
  L = []
  for i in range(len(T.data)):
    L = L + items_in_order(T.child[i])
    L.append(T.data[i])
  L = L + items_in_order(T.child[-1])
  return L
  
print('The items in the tree are',items_in_order(T))
```

## **Problem Type 3** - Function that traverses a single path that is dependent of parameters and tree contents 

```python
def items_in_ks_node(T,k):
  if k in T.data:
    return T.data
  if len(T.child)==0:
    return []
  ind = T.find_subtree(k)
  return  items_in_ks_node(T.child[ind],k)

for k in [10, 25, 30]:
  print('The items in the same node as',k,'are',items_in_ks_node(T,k))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

class BSTNode:
    def __init__(self, node_key):
        self.key = node_key
        self.left = None
        self.right = None
        self.parent = None

    def get_key(self):
        return self.key

    def get_parent(self):
        return self.parent

    def get_left(self):
        return self.left

    def get_right(self):
        return self.right

    def set_key(self, new_key):
        self.key = new_key

    def set_left(self, new_left_child):
        self.left = new_left_child
        if self.left:
            self.left.parent = self

    def set_parent_to_null(self):
        self.parent = None

    def set_right(self, new_right_child):
        self.right = new_right_child
        if self.right:
            self.right.parent = self

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def get_root(self):
        return self.root

    def in_order(self, visitor, current=None):
        return self.in_order_node(visitor, self.root)

    def in_order_node(self, visitor, current):
        if current:
            left = current.get_left()
            if left:
                if not self.in_order_node(visitor, left):
                    return False
            if not visitor.visit(current):
                return False
            right = current.get_right()
            if right:
                if not self.in_order_node(visitor, right):
                    return False
        return True

    def pre_order(self, visitor, current=None):
        return self.pre_order_node(visitor, self.root)

    def pre_order_node(self, visitor, current):
        if current:
            if not visitor.visit(current):
                return False
            left = current.get_left()
            if left:
                if not self.pre_order_node(visitor, left):
                    return False
            right = current.get_right()
            if right:
                if not self.pre_order_node(visitor, right):
                    return False
        return True

    def post_order(self, visitor, current=None):
        return self.post_order_node(visitor, self.root)

    def post_order_node(self, visitor, current):
        if current:
            left = current.get_left()
            if left:
                if not self.post_order_node(visitor, left):
                    return False
            right = current.get_right()
            if right:
                if not self.post_order_node(visitor, right):
                    return False
            if not visitor.visit(current):
                return False
        return True

    def breadth_first_traversal(self, visitor):
        if not self.root:
            return []

        visited = {}
        result = []
        queue = []

        current_node = self.root
        queue.append(current_node)

        while queue:
            current_node = queue.pop(0)
            result.append(current_node.get_key())
            visited[current_node] = True

            if current_node.get_left() and current_node.get_left() not in visited:
                queue.append(current_node.get_left())

            if current_node.get_right() and current_node.get_right() not in visited:
                queue.append(current_node.get_right())

        return result

    def find_min(self):
        current = self.root
        while current.get_left() is not None:
            current = current.get_left()
        return current

    def find_max(self):
        current = self.root
        while current.get_right() is not None:
            current = current.get_right()
        return current

    def insert_key(self, key):
        self.insert_node(self.make_new_node(key))

    def remove_key(self, key):
        node = self.search(key)
        if not node:
            return False
        else:
            return self.remove_node(node)

    def search(self, desired_key):
        current_node = self.root
        while current_node:
            if current_node.get_key() == desired_key:
                return current_node
            elif desired_key < current_node.get_key():
                current_node = current_node.get_left()
            else:
                current_node = current_node.get_right()
        return None

    def insert_node(self, node):
        if not self.root:
            self.root = node
        else:
            current_node = self.root
            while current_node:
                if node.get_key() < current_node.get_key():
                    if current_node.get_left() is None:
                        current_node.set_left(node)
                        current_node = None
                    else:
                        current_node = current_node.get_left()
                else:
                    if current_node.get_right() is None:
                        current_node.set_right(node)
                        current_node = None
                    else:
                        current_node = current_node.get_right()

    def make_new_node(self, key):
        return BSTNode(key)

    def remove_node(self, node_to_remove):
        if not node_to_remove:
            return False

        parent = node_to_remove.get_parent()

        if node_to_remove.get_left() and node_to_remove.get_right():
            successor_node = node_to_remove.get_right()
            while successor_node.get_left():
                successor_node = successor_node.get_left()

            node_to_remove.set_key(successor_node.get_key())
            self.remove_node(successor_node)
            return True

        elif node_to_remove == self.root:
            if node_to_remove.get_left():
                self.root = node_to_remove.get_left()
            else:
                self.root = node_to_remove.get_right()

            if self.root:
                self.root.set_parent_to_null()
            return True

        elif node_to_remove.get_left():
            parent.replace_child(node_to_remove, node_to_remove.get_left())
            node_to_remove.set_parent_to_null()

        elif node_to_remove.get_right():
            parent.replace_child(node_to_remove, node_to_remove.get_right())
            node_to_remove.set_parent_to_null()

        return True

class NodeVisitor:
    def __init__(self, bst):
        self.bst = bst

    def visit(self, node):
        # Implement this method to define what to do when visiting a node.
        # For example, you can print the node's key.
        print(node.get_key())

    def plot_in_order(self):
        nodes = []
        self.bst.in_order(lambda node: nodes.append(node.get_key()))
        plt.figure(figsize=(8, 4))
        plt.title("In-order Traversal")
        plt.plot(nodes, marker='o')
        plt.xlabel("Nodes")
        plt.ylabel("Keys")
        plt.grid(True)
        plt.show()

    def plot_breadth_first(self):
        nodes = self.bst.breadth_first_traversal(lambda node: node.get_key())
        plt.figure(figsize=(8, 4))
        plt.title("Breadth-first Traversal")
        plt.plot(nodes, marker='o')
        plt.xlabel("Nodes")
        plt.ylabel("Keys")
        plt.grid(True)
        plt.show()

    def plot_tree(self):
        G = nx.Graph()
        self._build_graph(G, self.bst.root)
        pos = nx.spring_layout(G)
        labels = {node: str(node.get_key()) for node in G.nodes()}
        nx.draw(G, pos, labels=labels, with_labels=True, node_size=500, node_color='skyblue', font_size=12)
        plt.title("Binary Search Tree")
        plt.show()

    def _build_graph(self, G, node):
        if node:
            if node.get_left():
                G.add_edge(node.get_key(), node.get_left().get_key())
                self._build_graph(G, node.get_left())
            if node.get_right():
                G.add_edge(node.get_key(), node.get_right().get_key())
                self._build_graph(G, node.get_right())

if __name__ == "__main__":
    bst = BinarySearchTree()
    keys_to_insert = [50, 30, 70, 20, 40, 60, 80]

    for key in keys_to_insert:
        bst.insert_key(key)

    print("In-order traversal:")
    visitor = NodeVisitor(bst)
    bst.in_order(visitor.visit)  # Call the visit method directly
    visitor.plot_in_order()

    print("\nPre-order traversal:")
    bst.pre_order(visitor)

    print("\nPost-order traversal:")
    bst.post_order(visitor)

    print("\nBreadth-first traversal:")
    print(bst.breadth_first_traversal(visitor))
    visitor.plot_breadth_first()

    search_key = 40
    found_node = bst.search(search_key)
    if found_node:
        print(f"\nKey {search_key} found in the tree.")
    else:
        print(f"\nKey {search_key} not found in the tree.")

    remove_key = 30
    if bst.remove_key(remove_key):
        print(f"\nKey {remove_key} removed from the tree.")
    else:
        print(f"\nKey {remove_key} not found in the tree.")

    print("\nIn-order traversal after removal:")
    bst.in_order(visitor)

    print("\nMinimum key:", bst.find_min().get_key())
    print("Maximum key:", bst.find_max().get_key())
```

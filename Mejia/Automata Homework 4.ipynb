```python
''''
Eric Quezada
CS 3350 Automata
Dr. Leobardo Valera
Homework 4, question 1:
Finitite Automaton Program
'''''
class FiniteAutomata:

    @staticmethod
    def simulate_finite_automaton(N, M, state, final_states, word):
        current_state = 0
        print("Simulating the finite automaton for the word '{}'".format(word))
        word = word.lower()

        for symbol in word:
            symbol_index = ord(symbol) - ord('a')

            if current_state < N and symbol_index < M:
                next_state = state[current_state][symbol_index]

                if next_state < N:
                    print("Current state: q{}, Symbol: {}, Next state: q{}".format(current_state, symbol, next_state))
                    current_state = next_state
                else:
                    print("Invalid state transition: q{} on symbol {}".format(current_state, symbol))
                    return False
            else:
                print("Invalid symbol or state: Symbol {}, Current state: q{}".format(symbol, current_state))
                return False

        if current_state < N and final_states[current_state]:
            print("Word '{}' is accepted! Current state q{} is a final state.".format(word, current_state))
            return True
        else:
            print("Word '{}' is not accepted. Ended in state q{}, which is not a final state.".format(word, current_state))
            return False


def main():
    print("Enter the number of states (N): ")
    N = int(input())

    print("Enter the number of symbols (M): ")
    M = int(input())

    state = [[0] * M for _ in range(N)]
    final_states = [False] * N

    print("Enter state transitions (state[N][M]):")
    for i in range(N):
        for j in range(M):
            print("state[{}][{}]: ".format(i, j), end="")
            state[i][j] = int(input())

    print("Enter final state information (final[N]):")
    for i in range(N):
        print("final[{}]: ".format(i), end="")
        final_states[i] = bool(input())

    print("Enter the word to test: ")
    word_to_test = input()

    FiniteAutomata.simulate_finite_automaton(N, M, state, final_states, word_to_test)


if __name__ == "__main__":
    main()
```

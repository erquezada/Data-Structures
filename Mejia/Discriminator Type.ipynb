```python
import os
import pandas as pd

# List all files in the current directory
files = os.listdir()

# Filter files that start with 'data' and end with '.csv'
data_files = [file for file in files if file.startswith('data') and file.endswith('.xlsx')]

# Iterate over each data file and process it
for file in data_files:
    # Read the excel file into a pandas DataFrame
    df = pd.read_excel(file)

    # Conditional assignment for 'classify' based on values in 'Response' column
    classify_conditions = ['statement', 'question', 'Statement', 'Question', 'higher', 'lower', 'Higher', 'Lower']
    df.loc[df['Response'].isin(classify_conditions), 'Question type'] = 'Classify'

    # Conditional assignment for 'obviousness' based on values in 'Response' column
    obviousness_conditions = ['Very', 'Not very', 'very', 'not very', 'obvious', 'not obvious', 'Obvious', 'Not Obvious']
    df.loc[df['Response'].isin(obviousness_conditions), 'Question type'] = 'Obviousness'

    # Save the updated DataFrame to a new excel file
    updated_file_name = f'updated_{file}'
    df.to_excel(updated_file_name, index=False)
```

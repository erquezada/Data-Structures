```python
import requests
from bs4 import BeautifulSoup

def scrape_video_names(url):
    # Fetch the webpage content
    response = requests.get(url)
    print("Response status code:", response.status_code)  # Debug statement
    if response.status_code == 200:
        # Parse the HTML content
        print("HTML content:", response.text)  # Debug statement
        soup = BeautifulSoup(response.text, 'html.parser')
        # Find elements containing video names
        video_name_elements = soup.find_all('h2', class_='g1-frame')  # Adjust this according to actual HTML structure
        # Extract video names
        video_names = [element.text.strip() for element in video_name_elements]
        return video_names
    else:
        print("Failed to fetch the webpage.")
        return []


# Example usage:
url = "https://sexycandidgirls.com/tag/upskirts/page/31"
video_names = scrape_video_names(url)
for name in video_names:
    print(name)
```

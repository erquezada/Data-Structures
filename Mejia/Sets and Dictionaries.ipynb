#Dictionaries

```python
student_grades = {
    "Alice": 85,
    "Amanda": 90,
    "Eric": 92
                 }

print(student_grades)
print(student_grades["Alice"])

student_grades["Tom"] = 92
print(student_grades)
student_grades.update({"Jessica": 98})
print(student_grades)
student_grades.pop("Amanda")
print(student_grades)
```

```python
inventory = {"apples": 10, "bananas": 25, "oranges" : 15}
keys = inventory.keys()
print(keys)
values = inventory.values()
print(values)
print(inventory.items())
```

```python
user_ages = {
    100: 25,
    101: 30,
    102: 24,
    103: 23
}

age = user_ages.get(101)
print(age)
age = user_ages.get(103, "Not found")
print(age)
```

```python
a = {"a": 1, "b": 2}
b = {"c": 3, "d": 4}

a.update(b)
print(a)
```

```python
person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

for key, value in person.items():
  print(f"{key}: {value}")
```

#Sets

```python
S = set(['red','blue','green'])
print(S)
print(len(S))

S2 = set()
print(S2)
print(len(S2))

S3 = set()
for i in range(5):
  S3.add(i)

print(S3)
print(len(S3))

S3.add(5)
print(S3)

S.union(S2)
print(S)
```

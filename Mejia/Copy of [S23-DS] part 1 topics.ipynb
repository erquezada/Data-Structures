# Objects and classes

```python
class My_class:

    def __init__(self, x=10, y=20):
        # a and b are the attributes of the object
        w = x*y      # w is a local varible in the initializer, it will be destroyed after execution
        self.a = x
        self.b = w

    # define a static method - to call it we do: My_class.my_static_method()
    def my_static_method():
        print('Running static method')
        # self is undefined in static methods

    # define a non-static method - to call it we do: object_name.my_non_static_method()
    # then self takes the value of object_name
    def my_non_static_method(self):
        print('Running non-static method')
        print('the object address is',self)
        print('the attribute values are: a =', self.a,'b =', self.b)

    def f(self, x, y): # if we call obj.f(10,20), then self = obj, x = 10, y = 20
        print(self, x, y)
```

```python
My_class.my_static_method()
```

```python
obj_1 = My_class(5)
obj_1.my_non_static_method()
```

```python
print('the object address is',obj_1)
print('the attribute values are: a =',obj_1.a,'b =', obj_1.b)
```

```python
obj_1.f(10,20)
```

# Sets

```python
S = set([4,1,8,9,2])
print(S)
```

```python
for s in S:
  print(s,end=' ')
```

```python
S = set()
for i in range(2,20,3):
  S.add(i)
  print(S)
```

```python
def foo3(nums): # n^2 - 10 + 10 log n => o(n^2 log(n)) => O(n^2)
  n = len(nums) # O(n)
  for i in range(n*n): # 0(n^2)
    if i >= 10: # if ten or more 0(1)
      print("UTEP") # print constant
    else: # n^2 - 10
      j = len(nums)
      while j > 0: # log n
        print("Miners")
        j //= 2
```

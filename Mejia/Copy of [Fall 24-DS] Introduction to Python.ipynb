### **for loops**





*for* loops are used for iterating over a sequence of items.

The statement below:

```
for i in seq:
  # loop body
```

will repeatedly execute the instructions in "# loop body", once for every element in variable *seq*. The variable *i* will take on the first value of  *seq* the first time through the loop, then the second, and so on

**The range function**

The most common way to generate a sequence of integers is through the *range* function.


```
range(start, stop, step)
```
returns a sequence of integers from start to stop with increments of size step.  

*range* receives 1, 2, or 3 parameters which specify the sequence start, end (non-inclusive), and step or increment, according to the following rules:

*   if 1 parameter is specified, it's assumed to be the end; start defaults to 0 and step to 1
*   if 2 parameters are specified, they're assumed to be start and end; step defaults to 1
*   if 3 parameters are specified, they're assumed to be start, end, and step

````python
for i in range(0,5,1):
    print(i)
````

````python
for i in range(1,10,2):
    print(i)
````

* start (optional): The starting value of the sequence. If not provided, it defaults to 0.
* stop (required): The end value of the sequence. The sequence generated will not include this value.
* step (optional): The increment between consecutive numbers in the sequence. If not provided, it defaults to 1.

Thus *range* may receive 1, 2, or 3 parameters:

* if 1 parameter is provided, it's assumed to be *stop*; *start* defaults to 0 and *step* defaults to 1
* if 2 parameters are provided, they're assumed to be *start* and *end*; *step* defaults to 1
* if 3 parameters are provided, they're assumed to be *start*, *end*, and *step*

````python
for i in range(10):
    print(i)
````

````python
for i in range(5,10):
    print(i)
````

````python
for i in range(1,10,2):
    print(i)
````

 *start*, *end*, and *step* can be negative.



````python
for i in range(-5,-1):
    print(i)
````

````python
for i in range(25,11,-1):
    print(i)
````

However, instead of using a negative step it is usually easier to use the **reversed** function to traverse a sequence backwards.

````python
for i in reversed(range(10)):
    print(i)
````

````python
for i in reversed(range(5,10)):
    print(i)
````

````python
for i in reversed(range(1,10,2)):
    print(i)
````

It is possible to have no elements in the specified *range*. This does not cause an error; the body of the loop is simply not executed

````python
for i in range(20, 10):
    print(i)
````

Notice that at the beginning of the loop, the loop variable (i) takes on the next value in the sequence, regardless of its current value.

````python
for i in range(3):
    for j in range(10,15):
        print(i,j)
````

## **Classes and objects**

Classes are similar to Java's


**Definition:** A class is defined using the **class** keyword followed by the class name.

**Attributes:** Classes can have attributes (data) and methods (functions).

**Constructor:** A special method called __init__() serves as the constructor in Python classes. It is called automatically when an object is created.

Thus classes in Python  are similar to classes in Java. However, unlike Java, Python does not have constructors with access modifiers like public, private or protected and does not allow multiple constructors.

Consider the following class in Java that allows to do the following:

implements a bank account :

*   Create a bank account with an initial balance
*   Deposit to an account
*   Withdraw from an account




```
public class BankAccount {
    private String accountNumber;
    private double balance;
    
    public BankAccount(String accountNumber, double balance) {
        this.accountNumber = accountNumber;
        this.balance = balance;
    }
    
    public void deposit(double amount) {
        if (amount > 0) {
            balance += amount;
            System.out.println("Deposit of $" + amount + " successful. New balance: $" + balance);
        } else {
            System.out.println("Invalid deposit amount");
        }
    }
    
    public void withdraw(double amount) {
        if (amount > 0 && amount <= balance) {
            balance -= amount;
            System.out.println("Withdrawal of $" + amount + " successful. New balance: $" + balance);
        } else {
            System.out.println("Insufficient funds or invalid withdrawal amount");
        }
    }
    
}
```



Now let's write the same class in Python.

````python
class BankAccount:
    def __init__(self, account_number, balance=0):  # Create object, store address in argument self
        self.account_number = account_number        # Create attribute 'account_number' and initialize to value of argument 'account_number'
        self.balance = balance                      # Create attribute 'balance' and initialize to value of argument 'balance'

    def deposit(self, amount):                      # self is the address of the object that the method is applied to; it is inderted automatically
        print(self)
        if amount > 0:
            self.balance += amount
            print(f"Deposit of ${amount} successful. New balance: ${self.balance}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):                     # self is the address of the object that the method is applied to; it is inderted automatically
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawal of ${amount} successful. New balance: ${self.balance}")
        else:
            print("Insufficient funds or invalid withdrawal amount")
````

The **self** parameter.

Here's what **self** represents and why it's used:

* **Reference to the Current Instance:** When you create an object (instance) of a class, self refers to that specific instance. It allows you to access the attributes and methods of the instance within the class.

* **Instance Method Binding:** When you call a method on an object (e.g., obj.method()), Python automatically passes the object itself as the first argument to the method. This allows the method to operate on the data specific to that instance.
* **Differentiating Instance Attributes:** self is used to differentiate instance attributes from local variables within methods. Without self, Python would treat variables created within methods as local variables, and they wouldn't be accessible outside of that method.

Thus all methods inside the class must have **self** as their first  parameter. **self** contains the address of the object. Python will automatically store the address of the object as the first parameter of a method call.



The method call

```
acct_1.deposit(100)
```
assigns the value of acct_1 to argument **self**, and 100 to argument **amount**.

Thus it would be equivalent to executing

```
BankAccount.deposit(acct_1,100)
```

To better understand what is being done, let's analyze what the following instruction does.

```
acct_1 = BankAccount(2302, 1000)
```

BankAccount(2302, 1000) calls the constructor of the **BankAccount** class, which will
* Allocate a block in memory to store an object
* Store the address of the new block in argument **self**
* Values 2302  and 1000 are stored in arguments **account_number** and  **balance** - recall that arguments are local variables and will disappear after executing the contructor.
* Create attributes **account_number** and  **balance** and initialize them to the values of arguments **account_number** and  **balance**
* Return the address of the new object (stored in argument **self**)

Then the address returned by the constructor is stored in variable **acct_1**

### **Notes on Style**

Code is read much more often than it is written, therefore writing code that is easy to read is very important.

Keeping a consistent style makes your code more readable. We strongly recommend using the guidelines described in https://peps.python.org/pep-0008/.

You do not need to memorize the guide, but remember to consult it when you have doubts about appropriate style.

Regarding naming, we recommend the following:

1.   Variable names should be descriptive and meaningful and describe what the variable represents or contains.
2.   Use **snake_case** for functions, methods, parameters, and variables.

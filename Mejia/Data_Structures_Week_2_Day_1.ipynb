```python
# Defining hello world function
def hello_world():
    print('Hello World')

# Defining for loop function for hello world
def multiple_hello_world():
    for i in range(10):
        print("Hello World")

# Defining for loop function
def print_evens():
    for i in range(0, 100, 2):
        print(i)

# Defining nested for loop function
def print_nested_evens():
    for i in range(0, 20, 2):
        for j in range(0, 10, 2):
            print(i, j)

# Defining main function
def main():
  user_option = int(input("Enter a number between 1 and 5: "))
  if user_option == 1:
    hello_world()
  elif user_option == 2:
    multiple_hello_world()
  elif user_option == 3:
    print_evens()
  elif user_option == 4 or user_option == 5:
    print_nested_evens()
  else:
      print("Please enter a valid number between 1 and 5.")

if __name__ == "__main__":
    main()
```

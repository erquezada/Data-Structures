```python
def calculate_cost_per_use(total_cost, number_of_uses):
    try:
        total_cost = float(total_cost)
        number_of_uses = int(number_of_uses)

        if number_of_uses <= 0:
            raise ValueError("Number of uses must be greater than zero.")

        cost_per_use = total_cost / number_of_uses
        return cost_per_use

    except ValueError as e:
        return str(e)

# Example usage:
total_cost_input = input("Enter the total cost: ")
number_of_uses_input = input("Enter the number of uses: ")

result = calculate_cost_per_use(total_cost_input, number_of_uses_input)

if isinstance(result, float):
    print(f"Cost per use: ${result:.2f}")
else:
    print(f"Error: {result}")
```

```python
def fact(n): # O(n^2)
  if n == 0:
    return 1
  else:
    return n * fact(n-1)

n = 7
print(fact(n))
```

```python
def fib(n): # O(2n!)
   if n <= 1:
       return n
   else:
       return(fib(n-1) + fib(n-2))

n = 8
print(fib(n))
```

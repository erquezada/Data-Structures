```python
import tkinter as tk
from tkinter import messagebox
```

```python

def print_grid(grid):
    for i in range(9):
        for j in range(9):
            entry_grid[i][j].delete(0, tk.END)
            entry_grid[i][j].insert(0, str(grid[i][j]))

def solve_sudoku(grid):
    for x in range(9):
        for y in range(9):
            if grid[x][y] == 0:
                for n in range(1, 10):
                    if is_valid(x, y, n, grid) and is_valid_neighbor(x - x % 3, y - y % 3, n, grid):
                        grid[x][y] = n
                        if solve_sudoku(grid):
                            return True
                        grid[x][y] = 0
                return False
    return True

def is_valid(x, y, n, grid):
    for i in range(9):
        if grid[x][i] == n or grid[i][y] == n:
            return False
    return True

def is_valid_neighbor(x, y, n, grid):
    for i in range(3):
        for j in range(3):
            if grid[x+i][y+j] == n:
                return False
    return True

def solve():
    for i in range(9):
        for j in range(9):
            if entry_grid[i][j].get() == '':
                grid[i][j] = 0
            else:
                grid[i][j] = int(entry_grid[i][j].get())

    if solve_sudoku(grid):
        print_grid(grid)
    else:
        messagebox.showinfo("Sudoku Solver", "No solution exists.")

root = tk.Tk()
root.title("Sudoku Solver")

grid = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

entry_grid = []

for i in range(9):
    row = []
    for j in range(9):
        entry = tk.Entry(root, width=4, font=('Helvetica', 24))
        entry.grid(row=i, column=j)
        row.append(entry)
    entry_grid.append(row)

solve_button = tk.Button(root, text="Solve Sudoku", command=solve)
solve_button.grid(row=9, columnspan=9)

root.mainloop()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.model_selection import LeaveOneOut
```

```python
path = 'EPISD.csv'
df = pd.read_csv(path, encoding='iso-8859-1')
# -Prepare Data-
preX = df

# Separate numbers and string
X_num = preX.select_dtypes(include=['int64', 'float64'])
X_str = preX.select_dtypes(include=['object'])


# Use one hot enode to convert string columns into numerical values

X_str_enc = pd.get_dummies(X_str)
X = pd.concat([X_num, X_str_enc], axis = 1)

X = X.drop(columns=['Average SAT'])
Y = df['Average SAT']

# Split the training and test sets

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 0)

# Generate plots to visualize data more clearly (SAT and Family Income)
plt.scatter(X_train['Average Family Income of Zip Code'], Y_train)
plt.title("EPISD Dataset")
plt.xlabel("Average Family Income in Zip Code")
plt.ylabel("Average SAT")

# Generate plots to visualize data more clearly
plt.scatter(X_train['Average Family Income of Zip Code'], Y_train)
plt.title("EPISD Dataset")
plt.xlabel("Average Family Income in Zip Code")
plt.ylabel("Average SAT")
plt.show()

# ML Model
model = LinearRegression()


# Perform 5- fold cross-validation on the training set

# Use Leave One Out to train the model in all but one data point
loo = LeaveOneOut()
scores = cross_val_score(model, X_train, Y_train, cv=loo, scoring='neg_mean_squared_error')

print("Cross Validation scores: ", scores)

# Use fit, in the rest of the training set
reg = model.fit(X_train, Y_train)
r_squared = model.score(X_test, Y_test)
print(r_squared)
pre = model.predict(X_test)
```

```python
import urllib.request
import json
from urllib.parse import quote

def will_go_outside():
    api_key = 'bae6d11c03261225d59f1fee5f6aae7b' # Replace with your OpenWeatherMap API key
    city_name = 'El Paso, US' # Replace with the name of the city you are interested in
    url = f'http://api.openweathermap.org/data/2.5/weather?q={quote(city_name)}&appid={api_key}'
    with urllib.request.urlopen(url) as response:
        if response.getcode() == 200:
            weather_data = json.loads(response.read())
            weather = weather_data['weather'][0]['main'].lower()
            if weather == 'clear':
                return 'yes'
            elif weather == 'clouds':
                return 'maybe'
            else:
                return 'no'
        else:
            print(f'Error: Could not retrieve weather data. Status code {response.getcode()}')
            return 'unknown'

print(will_go_outside())
```
